# Multivariate Analysis

```python
#I am drawing a pie chart to understand the percentage of deaths vs survival
'''The following links were helpful to understand on how to draw piecharts using matplotlib
   1)https://matplotlib.org/3.1.1/gallery/pie_and_polar_charts/pie_features.html#sphx-glr-gallery-pie-and-polar-charts-pie-features-py
   2)https://pythonspot.com/matplotlib-pie-chart/'''


#1 = the patient survived 5 years or longer 
#2 = the patient died within 5 year
labels = 'the patient survived 5 years or longer', 'the patient died within 5 year'
sizes = [224,81]
colors = ['yellow', 'lightblue']
explode = (0.1, 0)  # explode 1st slice

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')   # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

My Obeservations:

1)There are luckily no null values in the dataset

2)The chances of deaths were very less, which is nearly about 27% and comparatively the occurence of survival was relatively high.

```python
#Let's draw a pair plot to understand the relationship between various independent variables in the dataset

sns.set_style("whitegrid")
sns.pairplot(df,hue='survived',size=3)
plt.show()
```

1)Operation has been performed on diverse age groups

2)Most of the operations are performed in 1962 with std-dev 3 years

3)75% of patients have atmost 4 axil nodes

4)This can be the reason behind high survival rate of 0.73!

5)axil nodes are irregularly distributed. May have outliers; Or most of data is nearer to minimum value as mean is nearer to minimum value.


# Univariate Analysis

```python
'''lets use a special funtion from seaborn which is facetgrid...best explanation can be found on 
this link[https://www.tutorialspoint.com/seaborn/seaborn_facet_grid.htm]'''

plt.close();
sns.FacetGrid(df, hue="survived", size=5) \
   .map(sns.distplot, "age") \
   .add_legend();
plt.show();
```

My Observtion:

1) It looks like a Normal Distribution

2) Nothing much useful can be inferred from here. 

```python
plt.close();
sns.FacetGrid(df, hue="survived", size=5) \
   .map(sns.distplot, "op_year") \
   .add_legend();
plt.show();
```

My Observation: Still no useful information

```python
plt.close();
sns.FacetGrid(df, hue="survived", size=5) \
   .map(sns.distplot, "axil_nodes") \
   .add_legend();
plt.show();
```

My Observation: Axil Nodes is a very useful feature that help us to infer some new knowledge

```python
#Box Plots
plt.close();
plt.figure()

plt.figure(figsize=(10, 5))
sns.boxplot(x='survived',y='op_year', data=df)

plt.figure(figsize=(10, 5))
sns.boxplot(x='survived',y='age', data=df)

plt.figure(figsize=(10, 5))
sns.boxplot(x='survived',y='axil_nodes', data=df)


plt.show()
```
